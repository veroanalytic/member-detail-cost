```python

# Libraries

import random
import datetime
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
from pathlib import Path
# import uuid
import servicedt

# Change these variables to keep static cohort but with different cost and service_dt
cost = 15
days = timedelta(days=15)

# Export file variables
output_file = "Static_Cohort_Member_Detail_" + datetime.now().strftime("%Y%m%d%H%M%S") + ".txt"
output_dir = Path("C:/Users/jisbe/pyproject/member-detail-cost/files")  # specify where you want file to be exported to



# Read extracted file
static_cohort = pd.read_csv('C:/Users/jisbe/pyproject/member-detail-cost/files/Mockedup_Member_Detail_20220930213111_Static.txt', sep= '|')

# Changing date of service by 10 days
static_cohort['Service_DT'] = pd.to_datetime(static_cohort['Service_DT'])
static_cohort['Service_DT'] = static_cohort['Service_DT'] - days

# Changing cost by $1
static_cohort['Cost'] = round(static_cohort['Cost'] + cost, 2)

# Change FileBatch_ID to current time
static_cohort['FileBatch_ID'] = datetime.now().strftime("%Y%m%d%H%M%S")

# Output to CSV
output_dir.mkdir(parents=True, exist_ok=True)
static_cohort.to_csv(output_dir / output_file, sep='|', index = False)
```
